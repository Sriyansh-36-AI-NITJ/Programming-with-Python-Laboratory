a.	Write a program to calculate area of a triangle using Heron’s Formula.

```python
print(" Calculate Area of a Triangle using Heron's Formula ")

side_a = float(input("Enter the length of side a: ")) # Example: 5.0
side_b = float(input("Enter the length of side b: ")) # Example: 6.0
side_c = float(input("Enter the length of side c: ")) # Example: 7.0

if (side_a + side_b > side_c) and \
   (side_a + side_c > side_b) and \
   (side_b + side_c > side_a):

    semi_perimeter = (side_a + side_b + side_c) / 2

    area = (semi_perimeter * (semi_perimeter - side_a) * \
                       (semi_perimeter - side_b) * (semi_perimeter - side_c))**0.5

    print(f"\nThe semi-perimeter (s) is: {semi_perimeter:.2f}")
    print(f"The area of the triangle is: {area:.2f} square units")
else:
    print("\nError: The given side lengths do not form a valid triangle.")
    print("Please ensure that the sum of any two sides is greater than the third side.")
```

a.	Write a program to solve the expression: x = 10 + 2 * 3 ** 2 - 8 / 4 .

```python
print(f"Original expression: 10 + 2 * 3 ** 2 - 8 / 4")

# Step 1: Exponentiation (highest precedence)
step1 = 3 ** 2
print(f"After 3 ** 2: 10 + 2 * {step1} - 8 / 4")

# Step 2: Multiplication and Division (from left to right)
step2 = 2 * step1
print(f"After 2 * {step1}: 10 + {step2} - 8 / 4")

step3 = 8 / 4
print(f"After 8 / 4: 10 + {step2} - {step3}")

# Step 3: Addition and Subtraction (from left to right)
step4 = 10 + step2
print(f"After 10 + {step2}: {step4} - {step3}")

x = step4 - step3
print(f"After {step4} - {step3}: {x}")

print(f"\nThe final result of the expression is: {x}")
```

b.	Write a program to print the reverse of a number.

```python
print("Number Reversal Program")

try:
    original_number_str = input("Enter an integer: ")
    original_number = int(original_number_str)

    is_negative = original_number < 0
    num_abs = abs(original_number)

    num_str = str(num_abs)
    reversed_str = ""

    for digit in num_str:
        reversed_str = digit + reversed_str

    if original_number == 0:
        reversed_number = 0
    else:
        reversed_number = int(reversed_str)

    if is_negative:
        reversed_number *= -1

    print(f"The reverse of {original_number} is: {reversed_number}")

except ValueError:
    print("\nError: Invalid input. Please enter an integer.")
```

b.	Write a program to print the Fibonacci series using recursion.

```python
print("--- Fibonacci Series using Recursion ---")

def fibonacci_recursive(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

try:
    num_terms_str = input("Enter the number of terms for the Fibonacci series: ")
    num_terms = int(num_terms_str)

    if num_terms < 0:
        print("Please enter a non-negative integer.")
    else:
        print(f"\nFibonacci Series (first {num_terms} terms):")
        for i in range(num_terms):
            print(fibonacci_recursive(i), end=" ")
        print()

except ValueError:
    print("\nError: Invalid input. Please enter an integer.")
```

11.	File Handling

a.	Write a program that counts the number of tabs, spaces and newline characters in a file.

```python
print("--- File Character Counter ---")

file_name = "sample_text_file.txt"

# Create a dummy file with some tabs, spaces, and newlines
file_content = "This is a sample text.\tIt has some tabs.\nAnd also newlines.   With spaces."

try:
    with open(file_name, 'w') as file:
        file.write(file_content)
    print(f"Created '{file_name}' with sample content.")

    # Initialize counters
    tab_count = 0
    space_count = 0
    newline_count = 0

    # Open the file in read mode
    with open(file_name, 'r') as file:
        content = file.read()

        for char in content:
            if char == '\t':
                tab_count += 1
            elif char == ' ':
                space_count += 1
            elif char == '\n':
                newline_count += 1

    print(f"\nAnalysis of '{file_name}':")
    print(f"Tabs: {tab_count}")
    print(f"Spaces: {space_count}")
    print(f"Newlines: {newline_count}")

except IOError as e:
    print(f"An error occurred while handling the file: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
class Department:
    def __init__(self, name):
        self.name = name
        print(f"Department '{self.name}' created.")

    def display_department(self):
        print(f"Department: {self.name}")

class Course:
    def __init__(self, name, year, department):
        self.name = name
        self.year = year
        self.department = department # A course belongs to a department
        print(f"Course '{self.name}' ({self.year}) in {self.department.name} created.")

    def display_course(self):
        print(f"Course Name: {self.name}")
        print(f"Course Year: {self.year}")
        print(f"Course Department: {self.department.name}")

class Student:
    def __init__(self, name, rollno, year):
        self.name = name
        self.rollno = rollno
        self.year = year
        self.enrolled_course = None # Student can be enrolled in one course
        print(f"Student '{self.name}' (Roll No: {self.rollno}) created.")

    def enroll_in_course(self, course):
        if isinstance(course, Course):
            self.enrolled_course = course
            print(f"Student {self.name} enrolled in course '{course.name}' of Department '{course.department.name}'.")
        else:
            print(f"Error: Invalid course object provided for {self.name}.")

    def display_student_details(self):
        print(f"\nStudent Name: {self.name}")
        print(f"Roll Number: {self.rollno}")
        print(f"Academic Year: {self.year}")
        if self.enrolled_course:
            print("Enrolled Course Details:")
            self.enrolled_course.display_course()
        else:
            print("Not currently enrolled in any course.")

print("\n--- Creating Departments ---")
comp_dept = Department("Computer Science")
ece_dept = Department("Electronics and Communication")

print("\n--- Creating Courses ---")
python_course = Course("Advanced Python Programming", 2024, comp_dept)
java_course = Course("Enterprise Java Development", 2024, comp_dept)

print("\n--- Creating Students ---")
alice = Student("Alice Smith", "CS001", 2024)
bob = Student("Bob Johnson", "EC002", 2024)

print("\n--- Enrolling Students ---")
alice.enroll_in_course(python_course)
bob.enroll_in_course(java_course)

print("\n--- Displaying Student Details ---")
alice.display_student_details()
bob.display_student_details()
```

c.	Write a program that overloads the + operator on a class Student that has attributes name and marks.

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __str__(self):
        return f"Student Name: {self.name}, Marks: {self.marks}"

    def __add__(self, other):
        if isinstance(other, Student):
            combined_marks = self.marks + other.marks
            new_name = f"{self.name} & {other.name}"
            return Student(new_name, combined_marks)
        elif isinstance(other, (int, float)):
            return Student(self.name, self.marks + other)
        else:
            return NotImplemented

student1 = Student("Alice", 85)
student2 = Student("Bob", 92)
student3 = Student("Charlie", 78)

print("Original Students:")
print(student1)
print(student2)
print(student3)

print("\n--- Adding Two Student Objects ---")
combined_students_marks = student1 + student2
print(f"Result of student1 + student2: {combined_students_marks}")
print(f"Individual marks sum: {student1.marks + student2.marks}")

print("\n--- Adding a Number to a Student's Marks ---")
updated_student1_marks = student1 + 10
print(f"Result of student1 + 10: {updated_student1_marks}")
print(f"Alice's original marks: {student1.marks}")

final_marks_for_bob = student2 + 5.5
print(f"Result of student2 + 5.5: {final_marks_for_bob}")
```

d.	Write a program that validates name and age as entered by the user to determine whether the person can cast vote or not.

```python
print("--- Voting Eligibility Checker ---")

try:
    # Get the user's name
    name = input("Enter your name: ")

    # Get the user's age and convert it to an integer
    age_str = input("Enter your age: ")
    age = int(age_str)

    # Validate age: must be a non-negative number
    if age < 0:
        print(f"\n{name}, age cannot be negative. Please enter a valid age.")
    elif age >= 18:
        print(f"\nCongratulations, {name}! You are {age} years old and eligible to cast your vote.")
    else:
        years_to_wait = 18 - age
        print(f"\nSorry, {name}. You are {age} years old. You will be eligible to vote in {years_to_wait} year(s).")

except ValueError:
    print("\nError: Invalid age input. Please enter a whole number for your age.")
```
